```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud

import matplotlib 
from matplotlib import rc
rc('font', family='‪C:\Windows\Fonts\H2GTRE.TTF')
```

```python
path = r'C:\Users\user\Desktop\30104_wordcloud\메테오.txt'
path
```

```python
text=open(path, encoding = 'utf-8').read()
text
```

```python
img=r'C:\Users\user\Desktop\30104_wordcloud\little.png'
Image.open(img)
```

```python
lp_mask = np.array(Image.open(img))
lp_mask
```

```python
lp_wc=WordCloud(
    font_path='‪C:\Windows\Fonts\H2GTRE.TTF',
background_color='cyan',
mask=lp_mask,
max_font_size=50,
contour_width=5,
contour_color='yellow').generate(text)
```

```python
lp_wc.words_
```

```python
lp_wc.to_image()
```

```python
lp_wc.to_file('')
```
