### [문제]

####    subplot을 이용해서 1행 2열 구조 2개씩
####    4가지 색상 다르게
####    4가지 스타일 다르게
####    4가지 마커 서로 다르게
####    범례표시
####    x,y제목

## x=0부터 5까지 단위 1

### y1=x+1

### y2=x+2

### y3=x+3

### y4=x+4

```python
x=np.arange(0,5,1)
y1=x+1
y2=x+2
y3=x+3
y4=x+4

plt.subplot(1,2,1)
plt.plot(x,y1,'y--',marker='.')
plt.plot(x,y2,'r-',marker='o')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(['y1','y2'])

plt.subplot(1,2,2)
plt.plot(x,y3,'g',marker='^')
plt.plot(x,y4,'y',marker='*')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(['y3','y4'])
```
